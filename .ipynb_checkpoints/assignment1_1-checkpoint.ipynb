```python
def dfListTotal(df, list_id):
    biglist = []
    inner_df = pd.DataFrame()

    bar = progressbar.ProgressBar().start()
    for i in range(len(list_id)):
        #print(list_id[i])        

        inner_list = []
        for j in range(len(df)):    
            if df.iloc[j]["ID2"] == list_id[i]:
                inner_list.append(df.iloc[j][["Lab", "Value"]].values.tolist())
        inner_df = pd.DataFrame(inner_list)
        inner_df["index"] = list_id[i]
        inner_df = inner_df.set_index("index")
        inner_df = inner_df.drop_duplicates(0, keep="last")
        #inner_df.columns = [0, list_id[i]]

        df_transposed = inner_df.transpose()
        df_transposed.columns = df_transposed.iloc[0]
        df_transposed.set_index = list_id[i]
        df_transposed = df_transposed.drop(0)
        #print(inner_df)
        biglist.append(df_transposed)
        
        bar.update(i)
    bar.finish()
    
    return biglist
```

```python
list_hepatoma_df = dfListTotal(hepatoma, list_hepatoma_id)
list_hepatitis_df = dfListTotal(hepatitis, list_hepatitis_id)
```

```python
def dfLablist(dataframe, list_df):
    for i in range(len(list_lab)):
        df = dataframe[dataframe["Lab"] == list_lab[i]] # df for each Lab
        df = df.drop(["ID2", "Sex", "Birth", "Dx_date", "Lab"], axis=1) # remove Lab column
        df = df.drop_duplicates("Value") # remove identical values
        df = df.reset_index(drop=True)
        df.columns = ["Dx", list_lab[i]] # replace the column name into Lab name

        list_df.append(df)
```

```python
list_hepatitis = []
list_hepatoma = []
dfLablist(hepatitis, list_hepatitis)
dfLablist(hepatoma, list_hepatoma)

print(len(list_hepatitis))
print(len(list_hepatoma))
```
